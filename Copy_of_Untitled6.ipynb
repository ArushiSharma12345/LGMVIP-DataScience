Importing important libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Reading data

```python
data=pd.read_csv('/content/Iris.csv',index_col=0)
```

Getting information about the data

```python
data.head()
```

```python
#Checking null values
data.isnull().sum()
```

```python
#Checking number of rows and columns in dataset
data.shape
```

```python
#Overall stats of the dataset
data.describe()
```

Here Target variable is Species

```python
#Checking the unique values in target variable
data.Species.unique()
```

```python
#Splitting data into training and testing
from sklearn import tree
from sklearn.model_selection import train_test_split
X=data[['SepalLengthCm','SepalWidthCm' ,	'PetalLengthCm' ,	'PetalWidthCm']]
Y=data.Species.values
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

Creating the model

```python
model=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
#Training our classifier
model.fit(X_train,Y_train)
```

```python
#Prediction
Y_pred=model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(Y_test,Y_pred)
```

```python
#Evaluation Summary
print(classification_report(Y_test,Y_pred))
```

Visualization of Decision Tree

```python
cn=data["Species"].unique().tolist()
fig = plt.figure(figsize=(25,20))
tree.plot_tree(model, feature_names=data.columns[:-1],class_names=cn,filled=True)
```
